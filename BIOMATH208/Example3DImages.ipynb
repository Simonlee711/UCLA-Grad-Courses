```python
# this next line is useful when you run jupyter notebooks locally, 
# but it may not work on google drive
%matplotlib notebook

# numpy is a library is for working with regular arrays, and other scientific operations
import numpy as np

# matplotlib is a plotting library
import matplotlib.pyplot as plt

# nibabel is for reading neuroimages
import nibabel as nib
```

```python
# look at the size of this image, rows, columns, and slices
print(I.shape)
# I is a 3D array
# we can index indidual pixels like this
print(I[1,2,3])
# or we can access 2D sections by fixing one index and varying the other two (below)
```

```python
# let's display a sagittal slice
f,ax = plt.subplots()
Isag = np.rot90(I[I.shape[0]//2]) # I[i,:,:] equivalent to I[i] (matplotlib allows you to leave out trailing indices)
ax.imshow(Isag, cmap='gray')
ax.set_title('sagittal slice')
```

```python
# let's display a coronal slice
f,ax = plt.subplots()
Icor = np.rot90(I[:,I.shape[1]//2]) # equivalent to I[:,j,:] or I[:,j]
ax.imshow(Icor, cmap='gray')
ax.set_title('coronal slice')
# note left side of image is right side of person
# this convention (radiological convention) is what you'd expect to see if a person is facing you
```

```python
# let's display an axial slice
f,ax = plt.subplots()
Iaxi = np.rot90(I[:,:,I.shape[2]//2]) # 
ax.imshow(Iaxi, cmap='gray')
ax.set_title('axial slice')
# here radiological convention
# imagine we see a patient laying in bed on their back with their feet toward you
# top of the image is pointing toward nose (rostral)
# bottom is pointing toward back (caudal)
# left is on the right, and right is on the left

# where in space is the first pixel (8 choices)
# what direction in space is the second pixel (3 choices)
# what direction in space is the second row of pixels (2 choices)
# 48 possible choices for how to orient a 3D image stored as a regular 3D array
```

```python
# let's display a sequence of slices
# it is useful to specify vmin and vmax so they are all displayed on the same colorscale
vminmax = np.quantile(I.ravel(),(0.01,0.99))
vmin = vminmax[0]
vmax = vminmax[1]
# how many slices to display
n_slices = 20
n_rows = np.sqrt(n_slices).astype(int)
n_cols = np.ceil(n_slices/n_rows).astype(int)

slices = np.linspace(0,I.shape[0]-1,n_slices).astype(int)
f,ax = plt.subplots(n_rows,n_cols, sharex=True, sharey=True)
ax = ax.ravel()
for a,s in zip(ax,slices):
    a.imshow(np.rot90(I[s]), cmap='gray', vmin=vmin,vmax=vmax)
f.suptitle('Sequence of sagittal slices')
    
    
```

```python
# coronal
slices = np.linspace(0,I.shape[1]-1,n_slices).astype(int)
f,ax = plt.subplots(n_rows,n_cols, sharex=True, sharey=True)
ax = ax.ravel()
for a,s in zip(ax,slices):
    a.imshow(np.rot90(I[:,s]), cmap='gray', vmin=vmin,vmax=vmax)
f.suptitle('Sequence of coronal slices')
```

```python
# axial
slices = np.linspace(0,I.shape[2]-1,n_slices).astype(int)
f,ax = plt.subplots(n_rows,n_cols, sharex=True, sharey=True)
ax = ax.ravel()
for a,s in zip(ax,slices):
    a.imshow(np.rot90(I[:,:,s]), cmap='gray', vmin=vmin,vmax=vmax)
f.suptitle('Sequence of axial slices')
```

```python
# Let's look at projections
Ip = np.sum(I,axis=0)
f,ax = plt.subplots()
ax.imshow(np.rot90(Ip), cmap='gray')
ax.set_title('Projection')
```

```python
# and look at maximum intensity projections
Ip = np.max(I,axis=0)
f,ax = plt.subplots()
ax.imshow(np.rot90(Ip), cmap='gray')
ax.set_title('MIP')
```
